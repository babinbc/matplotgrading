```python
import numpy as np
import matplotlib.pyplot as plt
row = np.ones( (10,10,3) )
col = np.ones( (10,10,3) )
top_right = np.ones( (10,10,3) )
top_left = np.ones( (10,10,3) )
bottom_right = np.ones((10,10,3))
bottom_left = np.ones( (10,10,3) )

row[0:5,:,:] = 0
plt.xlabel("Row grading from 0 to 5")
plt.imshow(row)
plt.show()
```

```python
col[:,5:10,:]=0
plt.xlabel("Column grading from 0 to 5")
plt.imshow(col)
plt.show()
```

```python
top_right[0:5:,5:10,:]=0
plt.xlabel("Top Right")
plt.imshow(top_right)
plt.show()
```

```python
top_left[0:5,0:5,:]=0
plt.xlabel("Top Left")
plt.imshow(top_left)
plt.show()
```

```python
bottom_right[5:10, 5:10, :]=0
plt.xlabel("Bottom Right")
plt.imshow(bottom_right)
plt.show()
```

```python
bottom_left[5:10:,0:5,:]=0
plt.xlabel("Bottom Left")
plt.imshow(bottom_left)
plt.show()
```
