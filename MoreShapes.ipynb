```python
import numpy as np
import matplotlib.pyplot as plt
test1 = np.ones((11,11,3))

test1[10,:,:] = 0
test1[:,0:11:10,:] = 0
plt.xlabel("U Shape")
plt.imshow(test1)
plt.show()
```

```python
test2 = np.ones((13,17,3))
test2[0:6,0:6,:] = 0
test2[0:13:2,:,:] = 0
plt.xlabel("U.S. Flag")
plt.imshow(test2)
plt.show()
```
