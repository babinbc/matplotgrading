```python
import numpy as np
import matplotlib.pyplot as plt
tShape = np.ones((11,11,3))
reverseT = np.ones((11,11,3))

alternates = np.ones((11,11,3))
chessBoard = np.ones((8,8,3))

tShape[0:1,:,:] = 0
tShape[:,5:6,:] = 0
plt.xlabel("T Shape")
plt.imshow(tShape)
plt.show()
```

```python
reverseT[10:11,:,:] = 0
reverseT[:,5:6,:] = 0
plt.xlabel("Reverse T")
plt.imshow(reverseT)
plt.show()
```

```python
alternates[0:1,0:11:2,:] = 0
plt.xlabel("Alternates")
plt.imshow(alternates)
plt.show()
```

```python
chessBoard[0:8:2,0:8:2,:] = 0
chessBoard[1:8:2,1:8:2,:] = 0
plt.xlabel("Chess Board")
plt.imshow(chessBoard)
plt.show()
```
